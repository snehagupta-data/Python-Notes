## Practice Space for Numpy

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Data Structure : [resturent_id,2021,2022,2023,2024]
sales_data=np.array([
[1,150000,180000,220000,250000], # The Local Slice
[2,120000,140000,160000,190000], # Brown Crust
[3,200000,230000,260000,300000], # Pizza point
[4,180000,210000,240000,270000], # Pizza hub
[5,160000,185000,205000,230000]  # Street Crust
])
print(sales_data)
```

```python
print(" ========== Vasai Pizza Analysis ==========")
print("\nSales Data Shape :",sales_data.shape)
print("\nSales data for 1st 3 resturent :\n",sales_data[:3])
print("\nTotal Sales per Year :",np.sum(sales_data[:,1:],axis=0))
print("\nMinimum Sales per Store :",np.min(sales_data[:,1:],axis=1))
print("\nMaximum Sales per year :",np.max(sales_data[:,1:],axis=0))
print("\nAverage Sales per Store :",np.mean(sales_data[:,1:],axis=1))
print("\nCumulative Sales per Store :\n",np.cumsum(sales_data[:,1:],axis=1))

```

```python
stores_name=np.array(["The Local Slice","Brown Crust","Pizza point","Pizza hub","Street Crust"])
vectorized_upper=np.vectorize(str.upper)
print("Store name in Upper Case :",vectorized_upper(stores_name))
vectorized_length=np.vectorize(len)
print("Length of Store Name :",vectorized_length(stores_name))
```

```python
#Broacasting is mechanism that allows NumPy to perform arithmetic operations on arrays of different shapes and sizes.

#Calculate Monthly Average Sales
monthly_avg=np.round(sales_data[:,1:]/12,2)
print("Monthly Average Sales data :\n",monthly_avg)
```
