## Pandas Series
Pandas is an one - dimensional labled array capable for holding any data type.

```python
import numpy as np
import pandas as pd
```

```python
labels=['a','b','c']
my_list=[10,20,30]
arr=np.array([11,12,13])
dic = {'Name':'Sneha','Age':19,'Topic':'Pandas'}
```

```python
pd.Series(my_list,index=labels)
```

```python
pd.Series(arr,labels)
```

```python
pd.Series(dic) #Note:Here key is itself index if we try to give index manually it gives error !
```

## Pandas DataFrame
DataFrame is two-dimensional, tabular data structure with labeled rows and columns.

```python
#DataFrame using Dictionary
data={
    'Name':['Sneha','Ruhi','Kajal','Tanish','Rohit','Anshika'],
    'Age':[19,20,21,15,18,15],
    'Department':['DS','AI','ML','DL','DS','ML']
}

df=pd.DataFrame(data)
print(df)
```

```python
#DataFrame using Nested List
temp_data=[
    ['Sneha',19,'DS'],
    ['Tanish',20,'AI'],
    ['Aarohi',21,'ML']
]
df2=pd.DataFrame(temp_data,columns=['Name','Age','Department']) 
print(df2) #Note: Here default index for rows and column (0,1,2...)
```

```python
#Selection Column
print(df['Name'])
print("\n")
print(df[['Name','Department']])
```

```python
#Add Column
df['Location']=['Pune','Delhi','Mumbai','Pune','Delhi','Mumbai']
print(df)
```

```python
#Remove Column
#df.drop('Location') ''' It Gives Error axix is must !'''
df.drop('Location',axis=1) #It just show that how table looks after deletion on coulm it doesn't remove coulmn permamently !
df.drop('Location',axis=1,inplace=True)
df.drop(['Age','Department'],axis=1)
```

```python
#Remove Row
df.drop(0,axis=0,inplace=True) #Give index for row deletion if custom is given then use that otherwise default index is used !
df.drop([1,3]) #Here the default axix=0 for row 
```

```python
#Selecting Rows and Column in Data Frame
# loc vs iloc
```

```python
#Selecting subset of DataFrame
df.loc[[1,2]][['Name','Age']]
```

Conditional Selection

```python
#Select the peoples who can vote
df[df["Age"]>=18]
```

```python
#Select with age<15 and Department=ML
df[(df["Age"]<18) & (df["Department"]=="ML")]
```
