## Feature Extraction on Anime Dataset

```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv(r'anime.csv')
```

```python
df.head() #Starting 5 rows 
```

Create a New for No. of Episodes :

```python
df.loc[4]['Title'] #Here we can see episodes are written in Title it self
```

```python
def episode_extract(txt):
    check=False
    data=""
    for i in txt:
        if i == "(":
            check=True
            continue
        if i==")":
            break
        if check:
            data=data+i
    return data
```

```python
df["Episodes"]=df["Title"].apply(episode_extract)
df.head()
```

```python
df['Episodes']=df['Episodes'].str.replace(" eps","")
df.head()
```

```python
df["Episodes"]=df["Episodes"].astype(int)
df.head()
```

Make a New Column for TimeStamp :

```python
def extraction_time(txt):
    check=False
    data=""
    for i in range(len(txt)):
        if txt[i]==")":
            for j in range(i+1,i+20):
                data=data+txt[j]
            return data
```

```python
df['Total_Time']=df['Title'].apply(extraction_time)
```

```python
df.head()
```

```python
def calculate_months(date_range):
    month_map = {
        "Jan": 1, "Feb": 2, "Mar": 3, "Apr": 4,
        "May": 5, "Jun": 6, "Jul": 7, "Aug": 8,
        "Sep": 9, "Oct": 10, "Nov": 11, "Dec": 12
    }

    start, end = date_range.split(" - ")

    sm, sy = start.split()
    em, ey = end.split()

    sy, ey = int(sy), int(ey)

    total_months = (ey - sy) * 12 + (month_map[em] - month_map[sm]) + 1
    return total_months
```

```python
df['Total_Months'] = df['Total_Time'].apply(calculate_months)
```
