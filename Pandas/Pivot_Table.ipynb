## Pivot Table

```python
import pandas as pd

data = {
    "Date": [
        "2025-01-01","2025-01-01","2025-01-02","2025-01-02","2025-01-03",
        "2025-01-03","2025-01-04","2025-01-04","2025-01-05","2025-01-05"
    ],
    "Region": [
        "North","South","North","East","West",
        "South","East","North","West","South"
    ],
    "Product": [
        "Laptop","Mobile","Laptop","Tablet","Mobile",
        "Laptop","Tablet","Mobile","Laptop","Tablet"
    ],
    "Salesperson": [
        "Amit","Riya","Amit","Neha","Rahul",
        "Riya","Neha","Amit","Rahul","Neha"
    ],
    "Units_Sold": [2, 5, 1, 3, 4, 2, 2, 6, 1, 3],
    "Revenue": [120000, 75000, 60000, 45000, 60000, 120000, 30000, 90000, 60000, 45000]
}

df = pd.DataFrame(data)
df
```

```python
pd.pivot_table(df,values="Revenue",index="Region",columns="Product") #ByDefault It Calculate Mean (aggfnc='mean')
```

```python
pd.pivot_table(df,values="Units_Sold",index="Salesperson",columns="Region",aggfunc='max')
```

```python
pd.pivot_table(df,values="Revenue",index="Salesperson",aggfunc='sum')
```
