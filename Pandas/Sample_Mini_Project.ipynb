Which Country has the Highest Population ?

```python
df[df['population']==df['population'].max()]
```

Capital of Country with Highest Population ?

```python
df[df['population']==df['population'].max()]['capital_city']
```

Which Country has the Least Population ?

```python
df[df['population']==df['population'].min()]['country']
```

Find Top 5 Countries with Highest democtratic Score ?

```python
df[['country','democracy_score']].sort_values(by="democracy_score",ascending=False).head(5)
```

How Many Total Regions are there ?

```python
print(df['region'].value_counts().count())
```

How Many Countries lies in the Eastern Europe Region ?

```python
df[df['region'] == 'Eastern Europe']['country']
```

Who is the Leader of 2nd Most Populated Country ?

```python
df.sort_values(by='population',ascending=False)[['country','political_leader']].iloc[1]
```

How Many Countries are there whos political leaders are Unknow ?

```python
df[df['political_leader'].isna()]['country']
```

How Many Country has Republic in their Full Name ?

```python
count = 0

def counting_republic(txt):
    global count
    if 'republic' in str(txt).lower():
        count += 1
    return txt

df['country_long'] = df['country_long'].apply(counting_republic)

print(count)
```
